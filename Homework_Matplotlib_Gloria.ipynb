```python
 %matplotlib inline

import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import numpy as np
import seaborn as sns

city_data_to_load = "city_data.csv"
ride_data_to_load = "ride_data.csv"

city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)
```

```python
ride_df.head()
```

```python
combined_data = pd.merge(city_df, ride_df, how='left', on='city')
combined_data.head()
```

```python
dfmean = combined_data.groupby(["city","type"])
dfmeanUrban = combined_data.loc[combined_data["type"] == "Urban",:]
mean_fare_urban = dfmeanUrban.groupby(["city"])
mean_fare_urban = mean_fare_urban["fare"].mean()
mean_fare_urban

dfmeanrural = combined_data.loc[combined_data["type"] == "Rural",:]
mean_fare_rural = dfmeanrural.groupby(["city"])
mean_fare_rural = mean_fare_rural["fare"].mean()
mean_fare_rural

dfmeanSuburban = combined_data.loc[combined_data["type"] == "Suburban",:]
mean_fare_suburban = dfmeanSuburban.groupby(["city"])
mean_fare_suburban = mean_fare_suburban["fare"].mean()
mean_fare_suburban

# mean_fare_city = dfmean["fare"].mean()
# mean_fare_city
```

```python
dfmeanUrban = combined_data.loc[combined_data["type"] == "Urban",:]
driver_counts_urban = dfmeanUrban.groupby(["city"])
# driver_counts_urban = driver_counts_urban["driver_count"].sum()
driver_counts_urban = driver_counts_urban["driver_count"].mean()

dfmeanRural = combined_data.loc[combined_data["type"] == "Rural",:]
driver_counts_rural = dfmeanRural.groupby(["city"])
# driver_counts_rural= driver_counts_rural["driver_count"].sum()
driver_counts_rural = driver_counts_rural["driver_count"].mean()

dfmeanSuburban = combined_data.loc[combined_data["type"] == "Suburban",:]
driver_counts_suburban = dfmeanSuburban.groupby(["city"])
# driver_counts_suburban = driver_counts_suburban["driver_count"].sum()
driver_counts_suburban = driver_counts_suburban["driver_count"].mean()
driver_counts_suburban
```

```python
# count the number of rides, make sure it totals to 2375 which is should. CHECKED.

df = combined_data.groupby(["city","type"])
ride_counts = combined_data["ride_id"].count()
ride_counts

dfmeanUrban = combined_data.loc[combined_data["type"] == "Urban",:]
ride_counts_urban = dfmeanUrban.groupby(["city"])
ride_counts_urban = ride_counts_urban["ride_id"].count()
ride_counts_urban

dfmeanUrban = combined_data.loc[combined_data["type"] == "Rural",:]
ride_counts_rural = dfmeanUrban.groupby(["city"])
ride_counts_rural = ride_counts_rural ["ride_id"].count()
ride_counts_rural

dfmeanUrban = combined_data.loc[combined_data["type"] == "Suburban",:]
ride_counts_suburban = dfmeanUrban.groupby(["city"])
ride_counts_suburban = ride_counts_suburban ["ride_id"].count()
ride_counts_suburban
```

```python
urban = plt.scatter(ride_counts_urban, mean_fare_urban, edgecolors="black",
            s=driver_counts_urban*10, c="coral", alpha=0.8)
rural = plt.scatter(ride_counts_rural, mean_fare_rural, edgecolors="black",
            s=driver_counts_rural*10, c="gold", alpha=0.8)
suburban = plt.scatter(ride_counts_suburban, mean_fare_suburban, edgecolors="black",
            s=driver_counts_suburban*10, c="lightskyblue", alpha=0.8)


plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.grid()

plt.legend((urban, rural, suburban),
           ('Urban', 'Rural', 'Suburban'),
           scatterpoints=1,
           title="City Types",
           loc='best',
           fontsize=8)


plt.text(45, 40, 'Note: Bubble size correlates with driver count per city.', style='italic',
        bbox={'facecolor': 'white', 'alpha': 2, 'pad': 8})
plt.savefig("Pyber Ride Sharing Data.png")
plt.show()
```
